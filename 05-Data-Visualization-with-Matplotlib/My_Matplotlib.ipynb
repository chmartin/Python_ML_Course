```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = x ** 2
```

```python
x
```

```python
y
```

```python
#FUNCTIONAL
plt.plot(x,y,'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object Oriented
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes2.set_title('Embedded Plot')
axes1.set_title('Larger Plot')
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2)
axes[0].plot(x,y)
axes[0].set_title('0')
axes[1].plot(y,x)
axes[1].set_title('1')
plt.tight_layout
```

```python
fig = plt.figure(figsize=(8,2),dpi=100)
axe = fig.add_axes([0,0,1,1])
axe.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1, figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my_picture.png',dpi=200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x^2')
ax.set_title('My Title')
ax.set_ylabel('Y Label')
ax.set_xlabel('X Label')
ax.plot(x,x**3,label='x^3')
ax.legend(loc=10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8c00',linewidth=3,ls='--',marker='^',ms=10,markerfacecolor='green',markeredgewidth=3,markeredgecolor='yellow')
ax.plot(y,x,color='purple',lw=20,alpha=0.5)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=2,ls='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
